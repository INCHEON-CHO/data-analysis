```python
index_list = list(pd.unique(df1.iloc[:,13]))
```

```python
f_list2 = []
for i in range(len(f_list1)):
    for j in range(len(index_list)):
        if index_list[j] in f_list1[i]:
            f_list2.append(f_list1[i])
```

```python
f_list3 = []
for i in range(len(f_list2)):
    if f_list2[i][-13:-11]=='11' or f_list2[i][-13:-11]=='41':
        f_list3.append(f_list2[i])
```

```python
import fiona
mountain = []
lat_lng = []

for i in range(len(f_list3)):
    name = ""
    layer = fiona.open(line2+'/'+f_list3[i],layer='tracks')
    for j in range(len(list(layer.keys()))):
        name += f_list3[i].split('_')[1]+str(j)
        mountain.append(name)
        lat_lng.append(layer[j]['geometry']['coordinates'])
        name=""
```

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.io import shapereader
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER


def make_map(projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(9, 13),
                           subplot_kw=dict(projection=projection))
    gl = ax.gridlines(draw_labels=True)
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
import cartopy.io.img_tiles as cimgt

request = cimgt.OSM()

extent = [126.816644931, 126.817408155,
          37.5813028496, 37.5823769479]

fig, ax = make_map(projection=request.crs)
ax.set_extent(extent)

img = ax.add_image(request, 14)
s = ax.add_geometries(shp, ccrs.PlateCarree(),
                      facecolor='none',
                      edgecolor='crimson',
                      linewidth=2)
```
