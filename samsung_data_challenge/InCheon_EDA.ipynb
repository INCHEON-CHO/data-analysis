```python
# 각변수들 EDA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# csv 파일 읽어오기
test = pd.read_csv('test_kor.csv',encoding='CP949')
```

```python
# test 셋 정보 파악
test
```

```python
# test 셋 칼럼 저장
test_cols = test.columns
```

```python
# train 셋 읽기 test 셋에서 가지고 있는 컬럼들로만
train = pd.read_csv("Train_Death_Info.csv",encoding = "CP949", usecols=test_cols)
```

```python
# train 셋 상위 5개 정보 확인
train.head()
```

```python
# train 셋 test 셋 합치기
result = pd.concat([train,test])
```

```python
# 정보확인
result.iloc[[25037],:]
```

```python
# 정보확인 
result.tail()
```

```python
# 새로운열 생성(도,시 병합)
new_region = pd.concat([result.발생지시도,result.발생지시군구],axis=1,join='inner')
```

```python
# 인덱스 재설정
new_region = new_region.reset_index(drop=True)
```

```python
new_region.발생지시도[new_region.발생지시도.isna()]='NaN'
```

```python
new_region.발생지시군구[new_region.발생지시군구.isna()]='NaN'
```

```python
# 새로운 리스트에다가 '시-구' 형성 
new_list = []

for i in range(len(new_region)):
    new_list.append(new_region.iloc[i,0]+'-'+new_region.iloc[i,1])
```

```python
new_list = pd.Series(new_list)
```

```python
new_list
```

```python
# result 데이터 인덱스 리셋 
result = result.reset_index(drop=True)
```

```python
# result 데이터에 new_region 칼럼 추가
result['new_region'] = new_list
```
