```python
#test_kor 파일 위치에 따른 위도 경도 구하기

import pandas as pd
import numpy as np
import requests
```

```python
#열이름 영어로 교체

test.columns=['D_N',
 'DayofWeek',
 'Death_cnt',
 'Casualties_cnt',
 'Serious_cnt',
 'Slightly_cnt',
 'Injury_Report_cnt',
 'City_Do',
 'City_Gun_Gu',
 'Type_Large_Group',
 'Type_Middle_Group',
 'Violation_Law',
 'Road_Type_L',
 'Road_Type',
 'Dang1_L',
 'Dang2_L']
```

```python
#검색할 주소 예시
location = '영통동'
```

```python
#요청 주소(구글맵)
URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address=' + location
```

```python
#URL로 보낸 Request의 Response를 response 변수에 할당
response = requests.get(URL)
```

```python
# JSON 파싱
data = response.json()
```

```python
# lat, lon 추출
lat = data['results'][0]['geometry']['location']['lat']
lng = data['results'][0]['geometry']['location']['lng']
```

```python
# 도 구분으로 bool 타입으로 저장

bool_do = test.loc[:,'City_Do'].isna()
```

```python
bool_do[39]
```

```python
# 군구 구분으로 bool 타입으로 저장

bool_gun_gu = test.loc[:,'City_Gun_Gu'].isna()
```

```python
#행의 개수만큼 포문으로 위도 경도 list에 삽입
lat = []
lng = []
for i in range(len(test)):
    if bool_do[i] and bool_gun_gu[i]:
        lat.append(0)
        lng.append(0)
    elif bool_do[i] or bool_gun_gu[i]:
        location = test.City_Do[i]+'시'
        URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address=' + location

        response = requests.get(URL)

        data = response.json()

        lat.append(data['results'][0]['geometry']['location']['lat'])
        lng.append(data['results'][0]['geometry']['location']['lng'])
    else:
        location = test.City_Gun_Gu[i]
        URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address=' + location

        response = requests.get(URL)

        data = response.json()

        lat.append(data['results'][0]['geometry']['location']['lat'])
        lng.append(data['results'][0]['geometry']['location']['lng'])
```

```python
#확인
lat
```

```python
#확인
lng
```

```python
#위도 경도를 df 타입으로 저장
df_lat_lng = pd.DataFrame({'lat':lat,'lng':lng})
```

```python
df_lat_lng
```

```python
#test 데이터에 위도 경도 데이터를 추가
test = pd.concat([test,df_lat_lng],axis=1)
```
