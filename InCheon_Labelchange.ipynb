```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
# 패키지 임폴트 및 라벨인코더 임폴트
```

```python
trafic_data = pd.read_csv("Train_Death_Info.csv",encoding = 'CP949')
#한글 내용이 포함되어있는 csv 파일 읽기
```

```python
trafic_data.당사자종별_2당[trafic_data.당사자종별_2당.isna()]='없음'
# Na 값을 '없음'으로 대체
```

```python
trafic_data[trafic_data.당사자종별_2당_대분류.isna() | trafic_data.당사자종별_2당.isna()]
```

```python
columns_idx=trafic_data.columns
# 포문을 돌리기 위해 열들을 시리즈 형태로 만들어서 넣어줌
```

```python
for i in range(len(columns_idx)):
    if trafic_data.loc[:,columns_idx[i]].dtype=='O':
        le.fit(trafic_data.loc[:,columns_idx[i]])
        new_dn = pd.Series(le.transform(trafic_data.loc[:,columns_idx[i]]))
        trafic_data.loc[:,columns_idx[i]] = new_dn

#데이터 타입이 오브젝트 형태인 애들을 라벨인코더로 형변환을 해줌
```

```python
trafic_data['월']=trafic_data['발생년월일시'].astype(str).str.slice(4,6)
trafic_data['일']=trafic_data['발생년월일시'].astype(str).str.slice(6,8)
trafic_data['시']=trafic_data['발생년월일시'].astype(str).str.slice(8,10)
trafic_data=trafic_data[['발생년', '월','일','시', '발생분', '주야', '요일', '사망자수', '사상자수', '중상자수', '경상자수',
       '부상신고자수', '발생지시도', '발생지시군구', '사고유형_대분류', '사고유형_중분류', '사고유형', '법규위반_대분류',
       '법규위반', '도로형태_대분류', '도로형태', '당사자종별_1당_대분류', '당사자종별_1당', '당사자종별_2당_대분류',
       '당사자종별_2당', '발생위치X_UTMK', '발생위치Y_UTMK', '경도', '위도', ]]

#발생년월일시를 월,일,시 로 나눠주고, 새로운 열로 추가함
#열을 원하는 배열로 바꿈
```
